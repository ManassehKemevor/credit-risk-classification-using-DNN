```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import Sequential,layers
from tensorflow.keras import callbacks
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("loan_risk_multiclass_300.csv")
df.head()
```

```python
df.info()
```

```python
x=df.drop("RiskClass",axis=1)
y=df["RiskClass"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=30)
```

```python
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
model=Sequential([
    layers.Dense(64,activation="relu",input_shape=(7,)),
    layers.Dense(32,activation="relu"),
    layers.Dense(16,activation="relu"),
    layers.Dense(7,activation="softmax")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
history=model.fit(xtrain,
                  ytrain,
                  batch_size=32,
                  epochs=200,
                  validation_split=0.2,
                  callbacks=[callbacks.EarlyStopping(monitor="val_loss",patience=10)]
                 )
```

```python
history.history
```

```python
model.evaluate(xtrain,ytrain)
```

```python
model.evaluate(xtest,ytest)
```

```python
trainAcc=history.history["accuracy"]
trainVal=history.history["val_accuracy"]
trainLoss=history.history["loss"]
trainVaLLoss=history.history["val_loss"]
plt.plot(trainAcc,label="training accuracy")
plt.plot(trainVal,label="validation accuracy")
plt.plot(trainLoss,label="training loss")
plt.plot(trainVaLLoss,label="validation loss")
plt.legend(loc='upper right')
plt.show()
```
